```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
    
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']= movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies['genres']
```

```python
movies.head()
```

```python
newdf= movies[['movie_id','title','tags']]
```

```python
newdf.head()
```

```python
newdf['tags'] = newdf['tags'].apply(lambda x:" ".join(x))
```

```python
newdf.head()
```

```python
newdf['tags'][0]
```

```python
newdf['tags']= newdf['tags'].apply(lambda x:x.lower())
```

**text vectorization**
**bag of word technique**

*Bag of Words (BoW) is a popular technique for text vectorization in natural language processing (NLP). It's used to represent text documents as numerical vectors that machine learning algorithms can work with. The fundamental idea behind BoW is to capture the frequency of words in a document while disregarding the order and structure of the words.*

```python
!pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(newdf['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
nlps = PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(nlps.stem(i))
        
    return " ".join(y)
```

```python
newdf['tags'][0]
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron action adventure fantasy sciencefiction')
```

```python
newdf['tags'] = newdf['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movies_index = newdf[newdf['title'] == movie].index[0]
    distances = similarity[movies_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(newdf.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
